```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data.csv'
data = pd.read_csv(file_path)
```

```python
data_cleaned = data.drop(columns=[col for col in data.columns if "Unnamed" in col])

missing_data = data_cleaned.isnull().sum()
print("Missing Data in Each Column:")
print(missing_data)
```

```python
indicator_presence = data_cleaned.pivot_table(index='Company', columns=['Indicator', 'Year'], aggfunc='size', fill_value=0)

print("Tabulation of Indicators Available for Each Company:")
print(indicator_presence)
indicator_presence.to_csv('data_findings/indicator_presence.csv')
```

```python
print("Number of Indicators")
print(data_cleaned['Indicator'].nunique())
```

```python
# Drop duplicates based on Indicator, Company, Units, and Remarks columns
units_validation = data_cleaned.drop_duplicates(subset=['Indicator', 'Company', 'Units', 'Remarks'])
units_validation = units_validation.sort_values(by=['Indicator'])
units_validation = units_validation.reset_index()

# Select only the relevant columns for display
units_validation_display = units_validation[['Indicator', 'Company', 'Units', 'Remarks']]

# Display the cleaned and filtered rows in a presentable manner
print("Unique Rows Based on Indicator, Units, and Remarks:")
print(units_validation_display.to_string())
units_validation_display.to_csv('data_findings/indicators.csv')
```
